# Script to aggregate median sea level data
Resulting csv file is supposed to be used within tools like Datawrapper

```python
import pandas as pd

#read csv file, parse the 'date' column as datetime
df = pd.read_csv('global_msl.csv', parse_dates=['date'])

# extract year from 'date' column
df['year'] = df['date'].dt.year

# group data by year and calculate median for each year
df_agg = df.groupby('year', as_index=False).median()

# multiply by 1000 for mm value
df_agg['global_msl'] = df_agg['global_msl'] * 1000

print(df_agg.head())

# write the aggregated data to a new csv file
df_agg.to_csv('global_msl_agg.csv', index=False)
```
